```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the insurance data from insurance.csv
data = pd.read_csv('insurance.csv')

# Extracting relevant columns
X = data['age'].values.reshape(-1, 1)  # Features (age)
y = data['charges'].values  # Target variable (charges)

# Creating a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Insurance Costs vs. Age')
plt.legend()
plt.show()

# Fitting a linear regression model
model = LinearRegression()
model.fit(X, y)

# Making predictions
y_pred = model.predict(X)

# Plotting the scatter plot with the best-fit line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Best-fit line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Insurance Costs vs. Age with Best-fit Line')
plt.legend()
plt.show()
```
