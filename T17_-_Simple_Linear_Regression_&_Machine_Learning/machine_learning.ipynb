2. For each of the following case studies, determine whether it is appropriate to utilise regression or classification machine learning algorithms. Justify your answers:
* 2.1 Classifying emails as promotion or social based on their content and metadata.
* 2.2 Forecasting the stock price of a company based on historical data and market trends.
* 2.3 Sorting images of animals into different species based on their visual features.
* 2.4 Predicting the likelihood of a patient having a particular disease based on medical history and diagnostic test results.


2. Answer here
* 2.1 Classification algorithm as this involves assigning emails into two categories (promotion or social), which is a classification problem.The output is discrete, either promotion or social.
* 2.2 Regression algorithm. Stock predictions are made using price which is a continuous numeric value based on historical data and trends, making it a regression problem. Regression algorithms are used when the output variable is a continuous value.
* 2.3 Classification algorithms are suitable when the output variable is a category or a class label. In this case, visual features of images can be used as input for classification algorithms like conv olutional neural networks (CNNs) for image classification tasks.
* 2.4 Classification and regression both. This case involves predicting the probability of a patient having a disease , which is a classification problem. The output is binary (having the disease or not having the disease), making it approprioate for classification. however, logistic regression can be used to base predictions on other factors, such as medical history and test results.
